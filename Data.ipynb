# Text Splitting from documents

```python
## Reading from a pdf file
from langchain_community.document_loaders import PyPDFLoader

loader=PyPDFLoader('Archita_Resume_updated.pdf')
docs=loader.load()
docs
```

How to recursively split text by characters

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter=RecursiveCharacterTextSplitter(
    chunk_size=500,chunk_overlap=50
)

#Final_document=text_splitter.create_documents(docs)  # useful to text documents
Final_document=text_splitter.split_documents(docs) 

```

```python
print(Final_document[0])
```

This text splitter is the recommended one for generic text. It is parameterized by a list of characters. It tries to split on them in order until the chunks are small enough. The default list is ["\n\n","\n"," ",""]. This has the effect of trying to keep all paragraphs( and then sentences, and then words) together as long as possible, as those would generically seem to be the strongest semantically related pieces of text.

* How the text is split: by list of characters
* How the chunk size is measured: by number of characters

# How to split by character-character Text Splitter

This is the simplest method. This split based on a given chacter sequence, which defaults to "\n\n". Chunk length is measured by number of characters.

1. How the text is split: by single character separator.
2. How the chunk size is measured: by the number of characters

```python
from langchain_community.document_loaders import TextLoader

loader=TextLoader("speech.txt")
docs=loader.load()
docs
```

```python
from langchain_text_splitters import CharacterTextSplitter

text_splitter=CharacterTextSplitter(separator="\n",chunk_size=500,chunk_overlap=20)
text_splitter.split_documents(docs)
```

# How to split by HTML header

HTMLHeaderSplitter is a "structure-aware" chunker that splits text at the HTML element level and adds metadata for each header "relevent" to any given chunk. It can return chunks element by element or combine elements with the same metadata, with the objectives of (a) keeping related text grouped (more or less) semantically and (b) preserving context-rich information encoded in document structures. It can be used with other text splitters as part of a chunking pipeline.

```python
from langchain_text_splitters import HTMLHeaderTextSplitter
html_string="""
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8" />
  <meta name="viewport" content="width=device-width, initial-scale=1.0"/>
  <title>My First Webpage</title>
  <style>
    body {
      font-family: Arial, sans-serif;
      background-color: #f2f2f2;
      padding: 20px;
    }
    h1 {
      color: #333399;
    }
    img {
      width: 300px;
      border-radius: 10px;
    }
  </style>
</head>
<body>
  <h1>Welcome to My Webpage</h1>
  <h2>Welcome to My homepage</h2>
  <h3>Welcome to My Titlepage</h3>
  <p>This is a simple HTML page created as an example.</p>
  <img src="https://via.placeholder.com/300" alt="Placeholder Image">
</body>
</html>

"""

header_to_split_on=[
    ("h1","Header1"),("h2","Header2"),("h3","Header3")
]

html_splitter=HTMLHeaderTextSplitter(headers_to_split_on=header_to_split_on)
html_header_splits=html_splitter.split_text(html_string)
```

```python
html_header_splits
```

```python
url="https://comfyai.run/documentation/LLM_local_loader"
header_to_split_on=[
    ("h1","Header1"),("h2","Header2"),("h3","Header3")
]

html_splitter=HTMLHeaderTextSplitter(headers_to_split_on=header_to_split_on)
html_header_splits=html_splitter.split_text_from_url(url)
```

```python
import json
import requests

json_data=requests.get("https://api....../openapi.json")
```

```python
from langchain_text_splitters import RecursiveJsonSplitter

json_splitter=RecursiveJsonSplitter(max_chunk_size=300)
json_chunks=json_splitter.split_json(json_data)

for chunk in json_chunks[:3]:
    print(chunk)

## way2:the splitter can output documents

docs=json_splitter.create_documents(texts=[json_data])
```
